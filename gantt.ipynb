```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates

tasks = [
    {"Task": "Analyse des besoins et définition du problème", "Start": "2024-09-26", "End": "2024-10-09"},
    {"Task": "Traitement des données", "Start": "2024-10-03", "End": "2024-11-10"},
    {"Task": "Développement du modèle", "Start": "2024-10-28", "End": "2024-12-08"},
    {"Task": "Entraînement et optimisation des hyperparamètres", "Start": "2024-11-25", "End": "2025-01-20"},
    {"Task": "Évaluation et test", "Start": "2025-01-10", "End": "2025-02-20"},
    {"Task": "Rédaction du rapport", "Start": "2025-02-01", "End": "2025-03-15"},
]

df = pd.DataFrame(tasks)
df["Start"] = pd.to_datetime(df["Start"])
df["End"] = pd.to_datetime(df["End"])
df["Duration"] = (df["End"] - df["Start"]).dt.days

fig, ax = plt.subplots(figsize=(10, 5))
for i, task in enumerate(df.itertuples()):
    ax.barh(task.Task, task.Duration, left=task.Start, color="skyblue")
ax.invert_yaxis()

ax.xaxis.set_major_locator(mdates.DayLocator(interval=14))
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m-%d"))
plt.xticks(rotation=45)

plt.xlabel("Date")
plt.ylabel("Tâches")
plt.title("Diagramme de Gantt Prévisionnel")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()
```
